# Monthly revenues for Maine and Maryland

```python
# change transaction date to date time 
# extract month

Store_S["Transaction Date"] = pd.to_datetime(Store_S["Transaction Date"])
```

```python
# create column for month--
Store_S["Month"] = Store_S["Transaction Date"].dt.to_period("M")
```

```python
#Join sales and store detail tables

sales_merged = Store_S.merge(
    Store_D,
    on = "Store ID",
    how="left")
```

```python
sales_merged.head()
# show tables merged
```

```python
# show only Maine and maryland
# This will filter the sales_merged to only show maine and Maryland

sales_filtered = sales_merged[
    sales_merged["State"].isin(["Maine","Maryland"])]
```

```python
# Calculate monthly revenue totals
monthly_revenue = sales_filtered.groupby(
    ["State","Month"])["Sale Amount"].sum().reset_index()
```

```python
# sort by state then month
monthly_revenue = monthly_revenue.sort_values(by=["State","Month"])
```

```python
#limit scrolling 
monthly_revenue.head(5)
```

# Total Revenue over full data period for Maine and Maryland

```python
#using sales_filterd with maine and maryland get sum for group by
total_revenue = sales_filtered.groupby("State")["Sale Amount"].sum().reset_index()
```

```python
total_revenue = total_revenue.sort_values(by="Sale Amount",ascending=False)
# Sorting highest revenue state first
```

```python
total_revenue
```

Key Findings: Maryland outperformed Maine in overall revenue,Maryland outperformed Maine by roughly 510%.

# Question 3.) How would you rank the sales performance of each store in each sales territory? Which are the top-performing stores?

```python
# Group total sales by state,store id and store location
store_performance = sales_filtered.groupby(["State","Store ID","Store Location"])["Sale Amount"].sum().reset_index()
 
# Sort stores from highest to lowest sales
store_performance = store_performance.sort_values(by="Sale Amount",ascending=False)
```

```python
store_performance.style.hide(axis="index")
```

Key findings: Immediately we can see North Harford and germantown stores are a big impact for Marylands territory. North Harfod accounts for 
 approximately 76.05% of Marylands total revenue. This Single store alone generated about $6.83 million more revenue than the entire state of Maine
 combined, basicaly store 736 produced about roughly 463.88% of Maines Total revenue.

 Store 734 (Germantown): generated 584,675.92 in revenue which accounts for approximately 5.11% of Marylands total revenue.

# Question 5.)What is the number of transactions per month by product category in each assigned territory?What is total sales revenue per month by category? What might this tell you about the most popular products, and where could there be opportunity for growth?

part 1.) What is the number of transactions per month by product category in each assigned territory?

```python
# 2 seperate merges is critical for this question

# Merging products table with category table
# attaching category names to the products table/data
product_data = products.merge(
    Product_cat,
    on=["CategoryID","SubcategoryID"],how="left")
```

```python
# Merge sales data with product and category information
sales_products=sales_filtered.merge(
    product_data,
    on="Prod Num",
    how="left")
```

```python
# Count transactions by state,month, and category 
monthly_transactions = sales_products.groupby(["State","Month","Category"]).size().reset_index(name="Transaction Count")
```

```python
# sort results
month_transactions = monthly_transactions.sort_values(by=["State","Month","Transaction Count"],ascending=[True,True,False])
```

```python
# Show results without index for neater visuals
monthly_transactions.style.hide(axis="index")
monthly_transactions.head()
```

# Question 5.) Part 2. 

 What is the total sale revenue per month by category?

```python
# Calculate revenue totals by state,month, and category
monthly_category_revenue = sales_products.groupby(["State","Month","Category"])["Sale Amount"].sum().reset_index()
```

```python
# sort results
monthly_category_revenue = monthly_category_revenue.sort_values(by=["State","Month","Sale Amount"],ascending=[True,True,False])
```

```python
monthly_category_revenue.style.hide(axis="index")
monthly_category_revenue.head()
```

Key finding: The analysis above indicates that technology-related products were the strongest driver of monthly sales performance.
Categories such as books,art supplies and Apparel and Merchandise generated much lower revenue totals.

# Reference to question 3.) How would you rank the sales performance of each store in each sales territory? Which are the top-performing stores?

```python
# Remove stores exceeding 500000, and focus on stores below that amount.
stores_clean = store_performance[
    store_performance["Sale Amount"] < 500000]

# Separate Maine stores
maine_stores = stores_clean[
    stores_clean["State"] == "Maine"]

# Separate Maryland stores
maryland_stores = stores_clean[
    stores_clean["State"] == "Maryland"]

# Create figure size
plt.figure(figsize=(12,8))

# Create Maine bars in red
plt.barh(
    maine_stores["Store Location"],
    maine_stores["Sale Amount"],
    color="red",
    label="Maine")

# Create Maryland bars in blue
plt.barh(
    maryland_stores["Store Location"],
    maryland_stores["Sale Amount"],
    color="blue",
    label="Maryland"
)

# Highest sales at top
plt.gca().invert_yaxis()

# Add labels and title
plt.xlabel("Total Sales Revenue")
plt.ylabel("Store Location")
plt.title("Store Sales Performance by Territory without top 2 stores")

# Add legend
plt.legend()

# Add revenue labels inside bars
for index, value in enumerate(stores_clean["Sale Amount"]):
    plt.text(
        value - 70000,
        index,
        f'${value:,.0f}',
        va='center',
        fontsize=15,
        fontweight='bold',
        color='white')

# Adjust spacing
plt.tight_layout()
# ================================================================
#Line graph without top 2 stores
# MAINE DATA
maine_data = monthly_revenue[
    monthly_revenue["State"] == "Maine"]

# REMOVE TOP MARYLAND STORES
sales_no_top = sales_filtered[
    ~sales_filtered["Store Location"].isin(
        ["North Harford", "Germantown"]
    )]

# Filter Maryland data without top stores
maryland_no_top = sales_no_top[
    sales_no_top["State"] == "Maryland"]

# Recalculate Maryland monthly revenue
monthly_revenue_no_top = maryland_no_top.groupby(
    "Month"
)["Sale Amount"].sum().reset_index()

plt.figure(figsize=(14,6))

# Maine Data
plt.plot(
    maine_data["Month"].astype(str),
    maine_data["Sale Amount"],
    color="red",
    label="Maine")

# Maryland line
plt.plot(
    monthly_revenue_no_top["Month"].astype(str),
    monthly_revenue_no_top["Sale Amount"],
    color="blue",
    label="Maryland Without Top Stores")

# Add grid
plt.grid(True)

# Show fewer x-axis labels
plt.xticks(
    maine_data["Month"][::6].astype(str),
    rotation=0)

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.title("Monthly Revenue Comparison: Maine vs Maryland Without Top 2 Stores")

# Add legend
plt.legend()

# Fix spacing
plt.tight_layout()
```

After removing the two highest store location, which were Maryland Store 736 (North Harford) and Store 734 (Germantown) exceeding 500k in sales revenue, we will focus on stores below that amount. The remaining Chart shows a clearer comparison of the mid and lower performing stores across both territories. Even after removing the top performing store in Maryland , the remaining Maryland locations still show stronger and more consistent sales performance compared to several Maine locations,which aligns with the trends observed in the monthly Revenue line graphs. Which indicates the territory size is a big play in our analysis.

# Reference to question 2.)What is monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data?

```python
# Filter Maine data
maine_data = monthly_revenue[
    monthly_revenue["State"] == "Maine"]

# Filter Maryland data
maryland_data = monthly_revenue[
    monthly_revenue["State"] == "Maryland"]

# Create figure size
plt.figure(figsize=(14,6))

# Create Maine line
plt.plot(
    maine_data["Month"].astype(str),
    maine_data["Sale Amount"],
    color="red",
    label="Maine"
)

# Create Maryland line
plt.plot(
    maryland_data["Month"].astype(str),
    maryland_data["Sale Amount"],
    color="blue",
    label="Maryland"
)

# Add grid lines
plt.grid(True)

# Show fewer x-axis labels
plt.xticks(
    maine_data["Month"][::6].astype(str),
    rotation=0)

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.title("Monthly Revenue Comparison: Maine vs Maryland")

# Add legend
plt.legend()

# Fix spacing
plt.tight_layout()
```

# Line graph analysis

Maine demonstrated slower but steady revenue growth over time, with gradual increases enter the late 2024 and 2025,indicating that the terrritory is improving despite operating with a smaller scale than MaryLand.However the large gap between the two revenue lines suggests that Maines Lower performance may be more related to territory size and few store locations rather than completely sales operations.

# Reference Question 5.) 
# part 2.)What is the total sale revenue per month by category?

With the bar charts below we are able to see the categories per territory that need more attention in.


```python
# Group Maine revenue by category only
maine_category = monthly_category_revenue[
    monthly_category_revenue["State"] == "Maine"
]
maine_category = maine_category.groupby(
    "Category"
)["Sale Amount"].sum().reset_index()

# Create figure
plt.figure(figsize=(12,6))

# Create bar chart
bars = plt.bar(
    maine_category["Category"],
    maine_category["Sale Amount"],
    color="red"
)
# show grids
plt.grid(axis='y',linestyle='--',alpha=0.4,linewidth=1.5)
plt.gca().set_axisbelow(True)

# Rotate labels
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.title("Maine Revenue by Category")

# Add ONE revenue label per bar
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'${height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Adjust spacing
plt.tight_layout()

# ==========================================
# Group Maryland revenue by category only
maryland_category = monthly_category_revenue[
    monthly_category_revenue["State"] == "Maryland"]

maryland_category = maryland_category.groupby(
    "Category"
)["Sale Amount"].sum().reset_index()

# Create figure
plt.figure(figsize=(12,6))

# Create bar chart
bars = plt.bar(
    maryland_category["Category"],
    maryland_category["Sale Amount"],
    color="blue")

# show grids
plt.grid(axis='y',linestyle='--',alpha=0.4,linewidth=1.5)
plt.gca().set_axisbelow(True)

# Rotate labels
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.title("Maryland Revenue by Category")

# Add ONE revenue label per bar,to show exact revenue
for bar in bars:
    height = bar.get_height()

    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'${height:,.0f}',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold')
    
# Adjust spacing
plt.tight_layout()
```

The bar charts show that Maine still maintains strong performance within its leading categories, especially Technology and Accessories and Textbooks,
showing that customer demand is still present despite operating on a smaller scale.Continiuing  to strengthen Maines Highest performing categories through promotions,loyalty programs and product focused marketing campaigns.Maryland top category for Technology is almost 7x more than maine with all the stores included.
